# 2 - Esse exercício faremos por etapas:

B) Exclua nessa etapa as linhas das colunas com dados faltantes que você analisar que esse método é o ideal. Justifique o motivo de estar excluindo essas linhas.

```python
# Excluir as linhas onde os dados são críticos (por exemplo, gênero ou churn)
df_cleaned = df.dropna(subset=['Genero', 'Churn'])

# Verificar se as linhas foram removidas
df_cleaned.info()
```

C) Nessa etapa substitua pela média, mediana ou moda as linhas das colunas com valores faltantes que você analisou e entendeu que esse era o melhor método. Não se esqueça de justificar sua escolha.

**Dica variáveis numéricas:** Não se esqueça de verificar a distrbuição dos dados, média e mediana.

**Dica variáveis categóricas:** Caso precise substituir uma variável categórica utilize um gráfico de barras para verificar a que mais aparece e utilizar ela na substituição.
Ou você pode optar por excluir a coluna, mas justifique sua escolha.

Caso opte por substituir você pode utilizar a função fillna:
df['SUACOLUNA'].fillna("VALOR QUE SUBSTITUIRA O NULO", inplace=True)
Ela funcionará como a replace.

```python
# Certifique-se de que df_cleaned é uma cópia do DataFrame original
df_cleaned = df.copy() 

# Verificar as colunas do DataFrame
print(df_cleaned.columns)

# Renomear colunas para remover espaços em branco de forma segura
df_cleaned = df_cleaned.rename(columns=lambda x: x.strip())

# Tratar valores ausentes
if 'Pagamento_Mensal' in df_cleaned.columns:
    df_cleaned['Pagamento_Mensal'] = df_cleaned['Pagamento_Mensal'].fillna(df_cleaned['Pagamento_Mensal'].mean())
else:
    print("A coluna 'Pagamento_Mensal' não está no DataFrame.")

if 'Total_Pago' in df_cleaned.columns:
    df_cleaned['Total_Pago'] = df_cleaned['Total_Pago'].fillna(df_cleaned['Total_Pago'].median())
else:
    print("A coluna 'Total_Pago' não está no DataFrame.")

if 'PaymentMethod' in df_cleaned.columns:
    df_cleaned['PaymentMethod'] = df_cleaned['PaymentMethod'].fillna(df_cleaned['PaymentMethod'].mode()[0])
else:
    print("A coluna 'PaymentMethod' não está no DataFrame.")

# Exibir o DataFrame após tratamento de valores ausentes
print(df_cleaned)

```

# 3 - Verifique se encontramos valores digitados incorretamente, ou com letras maiusculas ou minusculas, ou até mesmo mesmos valores porém escritos de forma dirente e corrija.

```python
# Corrigir capitalização inconsistente
df_cleaned.loc[:, 'Servico_Internet'] = df_cleaned['Servico_Internet'].str.lower()

# Verificar e substituir valores mal formatados ou duplicados
df_cleaned.loc[:, 'PaymentMethod'] = df_cleaned['PaymentMethod'].replace('Bank transfer (automatic)', 'bank_transfer')

# Exibir o DataFrame após correção de valores
print(df_cleaned)

```
